# Задача

Есть сто пронумерованных закрытых дверей. Вы делаете 100 проходов около дверей и меняете состояние каждой двери, т.е. вы делаете закрытую дверь открытой и наоборот. После первого прохода Вы посещаете каждую 2-ую дверь (т.е. 2,4,6,8 и т.д.), на следующем проходе вы посещаете каждую 3-ю дверь и так далее. Необходимо сказать какие двери будут открыты, а какие закрыты после последнего прохода.

```python
import collections
import matplotlib.pyplot as plt
```

# Решение

```python
# словарь со значениями
lux_dict = {}
for i in range(101):
    # нули не нужны
    if  i != 0:
        lux = []
        for y in range(101):
            # нули не нужны
            if y != 0:
                # проверяю деление на ноль и добавляю прошедшие проверку числа в словарь для каждого из i
                lacmus = (i) % (y)
                if lacmus == 0:
                    lux.append(y)
            lux_dict[i] = lux
```

проверим для 50. изначально дверь закрыта

 50: [1, 2, 5, 10, 25, 50]
 
у 50 - 6 делителей без остатка- это проходы

делитель1 - 1 открыта

делитель2 - 2 закрыта

делитель3 - 5 открыта

делитель4 - 10 закрыта

делитель5 - 25 открыта

делитель6 - 50 закрыта

**Вывод - на чётном количестве делителей дверь остаётся закрыта, на нечётном - открыта**

```python
lux_len = {}
for key in lux_dict:
       list_len = 0
       lux_lux = []
       for item in lux_dict[key]:
           list_len += 1
       lux_lux.append(list_len)
       lux_len[key] = list_len
```

```python
# список цветов 
colors = []
for key in range(1, 101):
    if lux_len[key] % 2 == 0:
        colors.append('black')
    else:
        colors.append('blue')

# список координат
x_coords = []
y_coords = []
for i in range(10):
    for j in range(10):
        x_coords.append(i)
        y_coords.append(j)


fig, ax = plt.subplots()
ax.scatter(x_coords, y_coords, c=colors, s=100)


ax.set_xlim([-1, 10])
ax.set_ylim([-1, 10])
ax.set_xticks(range(10))
ax.set_yticks(range(10))
ax.set_xticklabels(range(0, 10))
ax.set_yticklabels(range(1, 11))
ax.set_aspect('equal')
ax.set_title('xxx')


plt.show()
```
